## Build a decision tree classifier to predict whether a customer will purchase a product or service based on their demographic and behavioral data. Use a dataset such as the Bank Marketing dataset from the UCI Machine Learning Repository.

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt
```

```python
 # delimiter=';' - for spliting the columns
dataset = pd.read_excel('C:/Users/hp/Downloads/bank additional.xlsx')
dataset
```

```python
dataset.info()
```

```python
data = pd.get_dummies(dataset, drop_first=True)
```

```python
X = data.drop('y_yes', axis=1)
y = data['y_yes']
```

product = loan, housing.
service = contact, campaign, poutcome.

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
clf = DecisionTreeClassifier(random_state=42)
clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)

# Evaluate the model
print("Accuracy:", accuracy_score(y_test, y_pred))
```

```python
print("Classification Report:\n", classification_report(y_test, y_pred))
```

```python
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
```

```python
plt.figure(figsize=(30,10))
plot_tree(clf, filled=True, feature_names=X.columns, class_names=['No', 'Yes'], rounded=True)
plt.title('Decision Tree')
plt.savefig('Decition Tree.png')
plt.show()
```
