## Create a bar chart or histogram to visualize the distribution of a categorical or continuous variable, such as the distribution of ages or genders in a population.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_excel('C:/Users/hp/Downloads/API2.xlsx')
dataset.head(3)
```

```python
dataset.info()
```

```python
dataset.isnull().sum()
```

```python
data = dataset.set_index('Country Name').iloc[:, 4:].T
data = data.apply(pd.to_numeric, errors='coerce').dropna(how='all')
total_population = data.sum(axis=1)
# Create a bar chart
plt.figure(figsize=(15, 6))
total_population.plot(kind='bar')
plt.xlabel('Year')
plt.ylabel('Total Population')
plt.title('Total Population by Year')
plt.show()
```
